```python
# imports
import matplotlib.pyplot as plt
import numpy as np
import cv2

# mostrar im√°genes y figuras inline
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = 13,13
```

```python
# Espacio de color RGB
img = cv2.imread('images/raspi_full.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# Se separan los canales. Cada canal queda en un valor 0-255 (Entre mas claro, mas representativo de ese canal)
img_r, img_g, img_b = img[..., 0], img[..., 1], img[..., 2]
tile = np.hstack([img_r, img_g, img_b])
plt.imshow(tile, cmap='gray')
```

```python
# Espacio de color LAB
# L ‚Äì Lightness ( Intensity ).
# a ‚Äì color component ranging from Green to Magenta.
# b ‚Äì color component ranging from Blue to Yellow.
img_LAB = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
img_L, img_A, img_B = img_LAB[..., 0], img_LAB[..., 1], img_LAB[..., 2]
tile = np.hstack([img_L, img_A, img_B])
plt.imshow(tile, cmap='gray')
```

```python
# Espacio de color YCrCb
# Y ‚Äì Luminance or Luma component obtained from RGB after gamma correction.
# Cr = R ‚Äì Y ( how far is the red component from Luma ).
# Cb = B ‚Äì Y ( how far is the blue component from Luma ).
img_YCrCb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
img_Y, img_Cr, img_Cb = img_YCrCb[..., 0], img_LAB[..., 1], img_LAB[..., 2]
tile = np.hstack([img_Y, img_Cr, img_Cb])
plt.imshow(tile, cmap='gray')
```

```python
# Espacio de color HSV
# H ‚Äì Hue ( Dominant Wavelength ).
# S ‚Äì Saturation ( Purity / shades of the color ).
# V ‚Äì Value ( Intensitimg_LAB = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
img_HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_H, img_S, img_V = img_HSV[..., 0], img_HSV[..., 1], img_HSV[..., 2]
tile = np.hstack([img_H, img_S, img_V])
plt.imshow(tile, cmap='gray')
```
