```python
# imports
import matplotlib.pyplot as plt
import numpy as np
import cv2

# mostrar imágenes y figuras inline
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = 20, 20
```

# Mostrar imágenes

```python
img = cv2.imread('data/doors.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
#plt.imshow(img)
```

# Separar los canales de la imágen

```python
img_r, img_g, img_b = img_rgb[..., 0], img_rgb[..., 1], img_rgb[..., 2]
tile = np.hstack([img_r, img_g, img_b])
plt.imshow(tile, cmap='gray')
```

# Difuminando bordes

```python
#Leo otra imágen con mas bordes
img = cv2.imread('data/gears.jpg')
```

## Filtro Box

```python
kernel = np.ones((10, 10), np.float32)
kernel /= kernel.sum()
res = cv2.filter2D(img, -1, kernel, cv2.BORDER_REPLICATE)
tile = np.hstack([img, res])
plt.imshow(cv2.cvtColor(tile, cv2.COLOR_BGR2RGB))
```

# Filtro Gausiano

```python
res = cv2.GaussianBlur(img, (0,0), 5, cv2.BORDER_REPLICATE)
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
tile = np.hstack([img, res])
plt.imshow(cv2.cvtColor(tile, cv2.COLOR_BGR2RGB))
```

## Respuesta al impulso

```python
delta = np.zeros((128, 128), dtype=np.float32) 
delta[64, 64] = 1. #Impulso
res = cv2.GaussianBlur(delta, (0,0), 25, cv2.BORDER_REPLICATE) #La respuesta al impulso me da la función de transf
plt.imshow(res, cmap='gray')
```

# Gradientes

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gx = cv2.Sobel(gray, cv2.CV_32F, 1, 0, ksize=5)
gy = cv2.Sobel(gray, cv2.CV_32F, 0, 1, ksize=5)
plt.imshow(np.hstack([gx, gy]), cmap='gray')
```
